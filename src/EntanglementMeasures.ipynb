```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import colors
from matplotlib import gridspec
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
mpl.rcParams['text.usetex'] = True
mpl.rcParams['text.latex.preamble'] = r'\usepackage{{amsmath}}'
```

```python
orange = ["#ff8c00"]
blue = ["#4173b3"]
green = ["#66cdaa"]
red = ["#e85c47"]
alpha = [0.5,0.4,0.3,0.2,0.1,0.0]
beta = [0.9,0.8,0.6,0.4,0.2,0.1]
for i,c in enumerate(alpha):
        orange.append(colors.get_alpha_hex(orange[0],beta[i]))
        blue.append(colors.get_alpha_hex(blue[0],beta[i]))
        green.append(colors.get_alpha_hex(green[0],beta[i]))
        red.append(colors.get_alpha_hex(red[0],beta[i]))
black="#000000"
gray="#cccccc"
orange[0]='#a05000'
blue[0]='#2B5080'
green[0]='#2B8050'
red[0]='#b82c47'
```

```python
#Load data files 
dataFile = '../data/spatEE_gs_19_19_09_+0.010_+100.000_xxx.dat'
data = np.loadtxt(dataFile)
U_19 = data[:,0]
S1_19 = data[:,2]
S1acc_19 = data[:,3]

dataFile = '../data/spatEE_gs_18_18_09_+0.010_+100.000_133.dat'
data = np.loadtxt(dataFile)
U_18 = data[:,0]
S1_18 = data[:,2]
S1acc_18 = data[:,3]

dataFile = '../data/spatEE_gs_17_17_08_+0.010_+100.000_133.dat'
data = np.loadtxt(dataFile)
U_17 = data[:,0]
S1_17 = data[:,2]
S1acc_17 = data[:,3]

dataFile = '../data/spatEE_gs_16_16_08_+0.010_+100.000_133.dat'
data = np.loadtxt(dataFile)
U_16 = data[:,0]
S1_16 = data[:,2]
S1acc_16 = data[:,3]

dataFile = '../data/spatEE_gs_15_15_07_+0.010_+100.000_133.dat'
data = np.loadtxt(dataFile)
U_15 = data[:,0]
S1_15 = data[:,2]
S1acc_15 = data[:,3]

dataFile = '../data/spatEE_gs_14_14_07_+0.010_+100.000_133.dat'
data = np.loadtxt(dataFile)
U_14 = data[:,0]
S1_14 = data[:,2]
S1acc_14 = data[:,3]

dataFile = '../data/spatEE_gs_13_13_06_+0.010_+100.000_133.dat'
data = np.loadtxt(dataFile)
U_13 = data[:,0]
S1_13 = data[:,2]
S1acc_13 = data[:,3]

dataFile = '../data/spatEE_gs_12_12_06_06_+0.010_+100.000_133.dat'
data = np.loadtxt(dataFile)
U_12 = data[:,0]
Sp1_12 = data[:,8]
```

```python
with plt.style.context('./aps.mplstyle'):
    #Create the figure
    factor=3
    fig = plt.figure()
    ax =fig.add_subplot(111)

    #Grey vertical line at transition point
    ax.axvline(x=3.3,color=gray, linestyle='--')   
    
    ax.semilogx(U_19, S1_19*1, 's',  label=None, markersize = 3, markerfacecolor = black, markeredgewidth = '0.25', color=black,zorder=19)
    ax.semilogx(U_19, S1acc_19*factor, 'o',  label=None, markersize = 3, markerfacecolor = black, markeredgewidth = '0.25', color=black,zorder=19)

    ax.semilogx(U_18, S1_18*1, 's',  label=r'$ S(\ell)$', markersize = 3, markerfacecolor = green[1], markeredgewidth = '0.25', color=green[0],zorder=18)
    ax.semilogx(U_18, S1acc_18*factor, 'o',  label=r'$3 \times S^{\rm{acc}}(\ell)$', markersize = 3, markerfacecolor = blue[1], markeredgewidth = '0.25', color=blue[0],zorder=18)

    ax.semilogx(U_17, S1acc_17*factor, 'o',  label=None, markersize = 3, markerfacecolor = blue[2], markeredgewidth = '0.25', color=blue[0],zorder=17)
    ax.semilogx(U_17, S1_17*1, 's',  label=None, markersize = 3, markerfacecolor = green[2], markeredgewidth = '0.25', color=green[0],zorder=17)

    ax.semilogx(U_16, S1acc_16*factor, 'o',  label=None, markersize = 3, markerfacecolor = blue[3], markeredgewidth = '0.25', color=blue[0],zorder=16)
    ax.semilogx(U_16, S1_16*1, 's',  label=None, markersize = 3, markerfacecolor = green[3], markeredgewidth = '0.25', color=green[0],zorder=16)

    ax.semilogx(U_15, S1acc_15*factor, 'o',  label=None, markersize = 3, markerfacecolor = blue[4], markeredgewidth = '0.25', color=blue[0],zorder=15)
    ax.semilogx(U_15, S1_15*1, 's',  label=None, markersize = 3, markerfacecolor = green[4], markeredgewidth = '0.25', color=green[0],zorder=15)

    ax.semilogx(U_14, S1acc_14*factor, 'o',  label=None, markersize = 3, markerfacecolor = blue[5], markeredgewidth = '0.25', color=blue[0],zorder=14)
    ax.semilogx(U_14, S1_14*1, 's',  label=None, markersize = 3, markerfacecolor = green[5], markeredgewidth = '0.25', color=green[0],zorder=14)

    ax.semilogx(U_13, S1acc_13*factor, 'o',  label=None, markersize = 3, markerfacecolor = blue[6], markeredgewidth = '0.25', color=blue[0],zorder=13)
    ax.semilogx(U_13, S1_13*1, 's',  label=None, markersize = 3, markerfacecolor = green[6], markeredgewidth = '0.25', color=green[0],zorder=13)
    
    lgnd = plt.legend(loc=(0.03,0.26), handlelength=0,handleheight=1.5,frameon=False)
    ax.text(0.15, 0.8, r'$L=N=\begin{matrix}19 \\ \uparrow\\13 \end{matrix}$')
    ax.text(25, 2.02, r'$\uparrow$')

    axp = ax.twiny()
    axp.semilogx(U_12, Sp1_12/factor, 'd',  label=r'$ \frac{1}{3}S(n=6)$', markersize = 3.5, markerfacecolor = orange[1], markeredgewidth = '0.25', color=red[0],zorder=20)
    axp.legend(loc=(0.69,0.50), handlelength=0,handleheight=1.5,title=r'$L=N=12$',frameon=False)
    axp.tick_params(axis='x', top="off",which='both',labeltop='off')
    ax.tick_params(axis='y', right="off",which='both',labeltop='off')

    ax.set_xlabel(r'$U$')
    ax.xaxis.set_label_coords(0.59  ,-.06)

    ax.set_ylim(-.05,2.33)
    axp.set_ylim(-.05,2.33)
    ax.set_xlim(U_18[0]-.0008, U_18[-1]+10)
    axp.set_xlim(U_12[0]-.0008, U_12[-1]+10)

    plt.savefig('../figures/EntanglementMeasures.pdf', transparent=False)    
    plt.savefig('../figures/EntanglementMeasures.svg', transparent=False)    
    plt.show
```
